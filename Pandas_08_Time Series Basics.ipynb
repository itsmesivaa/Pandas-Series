# Converting strings to datetime objects with pd.to_datetime()

```python
import pandas as pd
```

```python
temp = pd.read_csv('temp.csv')
```

```python
temp
```

```python
temp.datetime
```

```python
#Converting object datatype to datetime datatype
pd.to_datetime(temp['datetime'])
```

```python
#Setting up datetime column as index to temp dataframe
temp = temp.set_index(pd.to_datetime(temp['datetime']))
```

```python
#Dropping datetime object column from temp dataframe
temp.drop('datetime',axis=1)
```

```python
#Converting strings to datetime in pandas default format is YYYY MM DD 
pd.to_datetime('20-12-2023')
```

```python
#Parsing dates even when not formatted correctly
pd.to_datetime('20120123')
```

```python
#Parsing dates even when not formatted correctly
pd.to_datetime('2012/01/23')
```

```python
pd.to_datetime('2025 09 30')
```

```python
pd.to_datetime('1980 15th may')
```

```python
pd.to_datetime('may 2020 24th')
```

```python
x = pd.to_datetime(['2025 09 30','may 2020 24th','1980 15th may'])
```

```python
#Iterating throught datetime index
for datetime_x in x:
    print(datetime_x)
```

# Indexing and Slicing in TimeSeries

```python
import pandas as pd
```

```python
a = pd.read_csv('temp.csv',parse_dates = ['datetime'],index_col = 'datetime')
```

```python
a
```

```python
#Labeled based indexing
a.loc['2016-12-31 22:00:00']
```

```python

a.loc['2015-09-23']
```

```python
#timebased indexing on dataframe
a.loc['2015-09-23 06:00:00']
```

```python
#Day based slicing data from dataframe from range
a.loc['2015-09-23 06:00:00': '2015-09-25 06:00:00']
```

```python
#Slicing data from 2015-09-23 upto the end of the dataframe
a.loc['2015-09-23 06:00:00':]
```

```python
#Pandas handling differnt kinds of format in date column
a.loc['14FEBRUARY2016']
```

```python
a.loc[['2015-09-23 06:00:00','2016-02-14 13:00:00']]
```
