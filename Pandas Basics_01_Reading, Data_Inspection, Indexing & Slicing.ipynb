# Indexing and Slicing with iloc[]

```python
import pandas as pd
```

```python
summer = pd.read_csv('summer.csv',index_col='Athlete')
```

```python
summer
```

```python
summer.info()
```

# Selecting rows with iloc[] which means indexed based search

```python
summer.iloc[2]
```

# Parsing records indexed based from last row

```python
summer.iloc[-1]
```

```python
#Slicing first five rows from Dataframe
summer.iloc[0:5]
```

```python
#Slicing last five rows from Dataframe
summer.iloc[-5:]
```

```python
summer.iloc[0][4]
```

```python
summer.iloc[0,:3]
```

```python
#Slicing specified rows along with specified columns
summer.iloc[24:29,[2,3,5,7]]
```

# Label based indexing and slicing with loc[]

```python
summer
```

```python
summer.loc['CHOROPHAS, Efstathios']
```

```python
summer.loc["PHELPS, Michael"]
```

```python
summer.loc[["PHELPS, Michael",'LEWIS, Carl'],['Year','Event','Medal']]
```

```python
#Slicing customized columns only for all rows
summer.loc[:,['Year','Event','Medal']]
```
