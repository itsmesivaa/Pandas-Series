# Creating Columns Based on Other Columns

```python
import pandas as pd
```

```python
titanic = pd.read_csv('titanic.csv')
```

```python
titanic
```

```python
#Calculating Date of Birth detials for the passenger based on Titanic disaster year 1912
1912-titanic['age']
```

```python
#Creating new Column on Titanic Dataframe as Passenger Date of Birth
titanic['passenger_dob'] = 1912-titanic['age']
```

```python
titanic
```

```python
#Creating count of Relatives travelled with respective Passengers
titanic['passenger_relatives_count'] = titanic['sibsp'] + titanic['parch']
```

```python
titanic
```

```python
#Dropping exisitng columns sibsp - Sibling/Spouse and parch = Parent/Children
titanic.drop(columns = ['sibsp','parch'],inplace = True)
```

```python
titanic
```

```python
#Creating new column named Current ticket fare for titanic ship based on cumulative inflation adjustment rate 3,051.45%
#Ref_site: https://www.in2013dollars.com/us/inflation/1912?amount=1
# $1 is equivalent to $31.51

titanic['fare_2023_adjusted'] = titanic['fare'] * 31.51
```
